```python
import pandas as pd
import numpy as np
data={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,30000,2000,8000,45000],
    'Units Sold':[25,60,35,80,50,15]
}
df=pd.DataFrame(data)
df
```

```python
# Data Binning
bins=[0,10000,30000,60000]
labels=['Low','Medium','High']
df['Price Category']=pd.cut(df['Price'],bins=bins,labels=labels)
df
```

```python
# Data Binning for units sold
bins=[0,10,35,85]
labels=['Less sell','Moderate sell','High sell']
df['Sell Category']=pd.cut(df['Units Sold'],bins=bins,labels=labels)
df
```

```python
# Data Formatting
# Convert Data Type
print(df.dtypes)
df['Units Sold']=df['Price'].astype(float)
print(df.dtypes)
```

```python
# Convert Text to Uppercase
df['Product']=df['Product'].str.upper()
df
```

```python
# Rounding Values
df['Price']=df['Price'].round(2)
df
```

```python
# Addtional Operations
# Sort Data
sorted_df=df.sort_values(by='Price',ascending=False)
sorted_df
```

```python
# Display Unique Values
print(df['Price Category'].unique())
```

```python
# DataSet:Food Delivery Orders
import pandas as pd
import numpy as np
data={
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_value':[250,450,120,600,300,150,500,220],
    'Delivery_time':[30,45,25,60,35,20,50,28],
    'Distance_km':[5,8,3,10,6,2,9,4]
}
df=pd.DataFrame(data)
df
```

```python
# Binning Order Value
bins=[0,200,400,600]
labels=['Low','Medium','High']
df['Order_Category']=pd.cut(df['Order_value'],bins=bins,labels=labels)
df
```

```python
# Binning Delivery Time
bins2=[0,30,45,75]
labels2=['Fast','Medium','Slow']
df['Delivery_Category']=pd.cut(df['Delivery_time'],bins=bins2,labels=labels2)
df
```

```python
# Data Formatting Examples
#Convert Distance To Float
df['Distance_km']=df['Distance_km'].astype(float)
df
```

```python
# Convert Order_Category to uppercase
df['Order_Category']=df['Order_Category'].str.upper()
df
```

```python
# Unique categories
print(df['Order_Category'].unique())
df
```

```python
# Count of Categories
print(df['Order_Category'].value_counts())
df
```

```python
# Sorting
print(df.sort_values(by='Order_value'))
df
```
